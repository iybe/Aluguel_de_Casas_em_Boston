```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('housingOK.csv')
```

```python
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sns.set_palette("GnBu_d")
sns.set_style('whitegrid')
sns.pairplot(df)
```

```python
colunas = ["CRIM","ZN","INDUS","RM","AGE","DIS","RAD","TAX","PTRATIO","B","LSTAT"]
sns.pairplot(df, x_vars=["MEDV"], y_vars=colunas)
```

```python
X = df[["CRIM","RM","DIS","RAD","TAX","PTRATIO","B","LSTAT"]]
y = df["MEDV"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
predictions = lm.predict(X_test)
```

```python
plt.scatter(y_test,predictions)
```

```python
from sklearn import metrics
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

```python
from sklearn import linear_model
mr = linear_model.Ridge(alpha=7)
mr.fit(X_train,y_train)
predictions = mr.predict(X_test)
```

```python
plt.scatter(y_test,predictions)
```

```python
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

```python
reg = linear_model.Lasso(alpha=0.1)
reg.fit(X_train,y_train)
predictions = reg.predict(X_test)
```

```python
plt.scatter(y_test,predictions)
```

```python
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

```python
reg = linear_model.BayesianRidge()
reg.fit(X_train,y_train)
predictions = reg.predict(X_test)
```

```python
plt.scatter(y_test,predictions)
```

```python
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```
